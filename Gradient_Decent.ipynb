```
import numpy as np


def gradient_decent(x,y):
  m_curr = 1
  b_curr = 0
  n = len(x)
  iterations = 1000
  learning_rate = 0.1

  for i in range(iterations):
    y_pred = m_curr * x + b_curr
    cost = sum([val **2 for val in (y-y_pred)])
    md = -np.sum(x*(y-y_pred))
    bd = -np.sum(y-y_pred)
    m_curr = m_curr-learning_rate * md
    b_curr = b_curr-learning_rate * bd
    print("m =  {}, c =  {} ,iteration =  {}, cost {}".format(m_curr,b_curr,i, cost))
   
x = np.array([0.5,2.3,2.9])
y = np.array([1.4,1.9,3.2])
gradient_decent(x,y)
```

```
import numpy as np


def gradient_decent1(x,y):
  m_curr = 0
  b_curr = 0
  n = len(x)
  iterations = 1000
  learning_rate = 0.05

  for i in range(iterations):
    y_pred = m_curr * x + b_curr
    cost = (1/n) * sum([val **2 for val in (y-y_pred)])
    md = -(2/n) * np.sum(x*(y-y_pred))
    bd = -(2/n) * np.sum(y-y_pred)
    m_curr = m_curr-learning_rate * md
    b_curr = b_curr-learning_rate * bd
    print("m =  {}, c =  {} ,iteration =  {}, cost {}".format(m_curr,b_curr,i, cost))
   
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_decent1(x,y)
```
